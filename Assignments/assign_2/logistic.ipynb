### 1. Inspect and plot some portion of the training data using pandas. Segregate the training and testing data into two separate variables consisting of ‘feature values’ and corresponding ‘predictions’ (the prediction column is titled ‘RainTomorrow’ in the dataset). To simplify the problem a bit, clean the whole data by carrying out the following sub-tasks:

```python
weather = pd.read_csv("weather_train.csv")
weather.head()
```

#### (a) Convert the predictions in the binary format by using ‘1’ for ‘YES’ and ‘0’ for ‘NO’.

```python
weather = weather.replace({"RainTomorrow":{"Yes":1, "No":0}})
weather.head()
```

#### (b) Identify and drop the feature columns having datatype ‘object’.

```python
obj_dtype_columns = weather.select_dtypes(include=['object']).columns
obj_dtype_columns
```

```python
weather = weather.drop(columns=obj_dtype_columns, axis=1)
weather.head()
```

#### (c) Identify cells having ‘NaN’ or ‘NA’ values and replace them with mean values of their respective columns.

```python
weather.isna().sum()
```

```python
# filling NaN or NA with their respective mean
weather = weather.fillna(weather.mean())
weather.head()
```

```python
# after filling missing values with mean 
weather.isna().sum()
```

#### (d) ormalize all the feature values by scaling them between 0 and 1. The values in feature matrix ‘X’ can be normalized as:



```python
# taking out the RainTomorrow
rain_tom = weather["RainTomorrow"]
weather = weather.drop(columns="RainTomorrow")
```

```python
# calculating the mean of each column
means = weather.mean()
means
```

```python
maxs = weather.max()
mins = weather.min()
```

```python
# Mean Normalizing

# Subtracting the mean from each column
weather = weather - means

# Dividing each column by its max - min value
weather = weather / (maxs - mins)
```

```python
# adding RainTomorrow column
weather["RainTomorrow"] = rain_tom
weather.head()
```
